```python
ir = pd.read_csv("iris.csv")
os.getcwd()
```

```python
ir.head()
```

```python
sns.scatterplot(x="sepal.length", y="sepal.width", data = ir, hue="variety", style="variety")
```

```python
sns.scatterplot(x="sepal.length", y="sepal.width", data = ir, hue="sepal.width", style = "variety")
```

```python
# histogram
```

```python
sns.displot(ir["sepal.length"], kde=True, color="Orange")
# log transformation
import math
ld = [math.log(d) for d in ir["sepal.length"]]
# plot
sns.displot(ld,bins = 50,kde=True)
```

```python
sns.displot(ir["sepal.length"], kde=True, color="blue", bins=18, rug=True)
```

```python
# joinplot
sns.jointplot(x="sepal.length", y="petal.length", data=ir, color="green", kind='reg')
```
